## 1 - Inputs and modeling setup

Efficiency curves, energy prices, and interest rate define the model inputs.  
Daily discharge and elevation scenarios are filtered in time, leap days removed, and scenarios indexed for simulation.

```python
import numpy as np 
import matplotlib.pyplot as plt
import pickle
from ptreeopt.plotting import *
from JP_plot_functions import *
import pandas as pd
from pathlib import Path

# from HydropoweerModel.ResSimModel import energy_model, Invar_sim
from mRun_HydropowerModel.mRun_ResSimModel import energy_model

# ---------------------------
# Define data inputs

# Efficiency curve file (text file with table of efficiency points)
Effcurve = np.loadtxt('YZeffcurves.txt')      

# ---------------------------
# Financial parameters
# Energy price tiers in $/kWh (midday, off-peak, etc.)
energy_price = np.array([0.09, 0.05, 0.03])

# Annual interest rate (decimal)
i_rate = 0.03

INPUT_DIR = Path("INput")

# ---------------------------
# Load future scenarious

daily_pre_discharges = pd.read_csv( INPUT_DIR / "Daily_predict_discharges.csv", parse_dates=["Date"])
daily_pre_elevations = pd.read_csv(INPUT_DIR / "Daily_predict_elevations.csv", parse_dates=["Date"])

# ---------------------------
# Step 1: filter from start_date onward
start_date = pd.Timestamp("1981-01-01")
daily_pre_discharges = daily_pre_discharges[daily_pre_discharges["Date"] >= start_date].copy()
daily_pre_elevations = daily_pre_elevations[daily_pre_elevations["Date"] >= start_date].copy()

# ---------------------------
# Step 2: drop Feb 29 (leap days)
# Create mask and apply it to each dataframe separately
mask_not_feb29 = ~((daily_pre_discharges["Date"].dt.month == 2) &
                   (daily_pre_discharges["Date"].dt.day == 29))
daily_pre_discharges = daily_pre_discharges[mask_not_feb29]

mask_not_feb29 = ~((daily_pre_elevations["Date"].dt.month == 2) &
                   (daily_pre_elevations["Date"].dt.day == 29))
daily_pre_elevations = daily_pre_elevations[mask_not_feb29]

# Reset index after filtering to keep integer indices for iloc
daily_pre_discharges.reset_index(drop=True, inplace=True)
daily_pre_elevations.reset_index(drop=True, inplace=True)

# ---------------------------
# Scenario settings and indexing
# total scenarios expected in the CSVs (excluding date column)
n_scenarios = 56

# all indices for scenario columns (shifted by +1 to skip the first column = date)
all_idx = np.arange(1, n_scenarios + 1) 

# Load indices to exclude (random samples) and compute the remaining set
random_idx = np.loadtxt('random_idx.txt', dtype=int)   # expects integers in file
idx = np.setdiff1d(all_idx, random_idx)               # columns to keep

# ---------------------------
# Extract scenario matrices
# iloc columns: 0 = Date, 1..n_scenarios = scenarios
sen_discharges = daily_pre_discharges.iloc[:, all_idx].to_numpy()
sen_elevations = daily_pre_elevations.iloc[:, all_idx].to_numpy()
```

## 2 - Simulation with optimal decision tree

The optimal decision tree is loaded and used to run the hydropower simulations across all scenarios.  
Model outputs are then analyzed and the decision tree is visualized.

```python
# ---------------------------
# Load optimal decision tree

# Load the optimized decision tree (policy)
snapshots = pickle.load(open("snapshots_optimal.pkl", "rb"), encoding="latin1")

# Assign decision tree to model input variable
P = snapshots

# ---------------------------
# Run hydropower simulation
# Initialize the hydropower model with physical, economic, and scenario inputs
model_V = energy_model( Effcurve, energy_price, i_rate, sen_discharges, sen_elevations)

# Simulate system performance under the optimal decision tree
Sim_of, Sim_DailyPowers, Sim_Xpar, Sim_Replacement_year, Sim_eff, Sim_head, \
Sim_Q, Sim_head_all, Qdesign, IC_added, Sim_PeakPowers = \
    model_V.f(P, mode='simulation')

# ---------------------------
# Visualize decision tree
# Plot the structure of the optimal decision tree
DecisonTree_Plot(P)
```

## 3 - Representative reservoir elevation scenarios
The driest and wettest scenarios are identified based on minimum and maximum mean elevation and plotted for comparison.

```python

# Identify representative elevation scenarios

# Compute mean reservoir elevation for each scenario
RL_means = np.mean(sen_elevations, axis=0)

idx1 = np.argmin(RL_means)   # driest (lowest mean elevation)
idx2 = np.argmax(RL_means)   # wettest (highest mean elevation)

# Extract date vector for plotting
date = daily_pre_discharges["Date"]

# Load precomputed daily power series for benchmark strategies

Cal_DailyPowers1 = np.loadtxt(INPUT_DIR / "ALL1R_Cal_DailyPowers.csv", delimiter=",")  # one replacement (R1)
Cal_DailyPowers2 = np.loadtxt(INPUT_DIR / "ALL2R_Cal_DailyPowers.csv",   delimiter=",")  # two replacements (R2)

# ---------------------------
# Plot reservoir elevation time series for dry and wet scenarios
EL_scenario_plots(sen_elevations, date, scenarios={idx1: "Dry Scenario", idx2: "Wet Scenario"})
```

## 4 - Parallel-plot input preparation: performance and climate metrics

```python
# ---------------------------
# Performance and climate metrics (organized, commented)

# Climate summary
# drought_indices returns (drought_freq, other_metrics, ).
# Threshold 300 uses the same units as sen_discharges.
drought_freq, _, _ = drought_indices(sen_discharges, 300)   # (n_scen,)
# 10th-percentile reservoir level per scenario (low-elevation indicator)
RL10 = np.quantile(sen_elevations, 0.1, axis=0)              # (n_scen,)
# Coefficient of variation of reservoir elevation per scenario
RL_cv = np.std(sen_elevations, axis=0) / np.mean(sen_elevations, axis=0)
# Mean discharge per scenario
D_means = np.mean(sen_discharges, axis=0)

# ---------------------------
# Load benchmark power series and NPVs
# Files expected: rows = days, columns = scenarios

NoR_DailyPowers = np.loadtxt(INPUT_DIR / "ALL_noR_Cal_DailyPowers.csv", delimiter=",")  # no replacement
#Cal_DailyPowers1 = np.loadtxt(INPUT_DIR / "ALL1R_Cal_DailyPowers.csv", delimiter=",")  # one replacement (R1)
#Cal_DailyPowers2 = np.loadtxt(INPUT_DIR / "ALL2R_Cal_DailyPowers.csv",   delimiter=",")  # two replacements (R2)

# Net present value arrays for the fixed replacement strategies

NPV1 = np.loadtxt(INPUT_DIR /"ALL1R_NPV.csv", delimiter=",")  # R1 NPV per scenario
NPV2 = np.loadtxt(INPUT_DIR /"ALL2R_NPV.csv", delimiter=",")  # R2 NPV per scenario

# ---------------------------
# Energy difference: HyTUNE vs fixed benchmarks
# Convert daily differences to annual GWh per scenario:
# mean(daily diff) * 365 / 1e6  => GWh 
dE1 = Sim_DailyPowers - Cal_DailyPowers1
dE1_mean = dE1.mean(axis=0) * 365 / 1e6

dE2 = Sim_DailyPowers - Cal_DailyPowers2
dE2_mean = dE2.mean(axis=0) * 365 / 1e6

# ---------------------------
# Installed power and capacity (model-specific scaling)
D_discharge = np.array([45] + [96.3] * 15 + [50])         # turbine discharge layout
# Installed power, using model constants and scaling to GW (1e6)
in_power = 150 * np.sum(D_discharge) * 9.81 / 1e6       # nominal installed power

# Example benchmark total capacities
TP_ex1 = in_power * 2
TP_ex2 = in_power * 3

# ---------------------------
# Mean annual energy per scenario (GWh) and capacity-normalized metrics
TE_cal1 = np.mean(Cal_DailyPowers1, axis=0) * 365 / 1e6     # R1 annual mean (GWh)
TE_cal2 = np.mean(Cal_DailyPowers2, axis=0) * 365 / 1e6     # R2 annual mean (GWh)
TE_sims = np.mean(Sim_DailyPowers, axis=0) * 365 / 1e6      # HyTUNE annual mean (GWh)

# Total installed capacity for HyTUNE per scenario (MW -> scale to same units as TE)
TP_sims = (np.sum(IC_added, axis=0) + in_power) / 1e6       # keep consistent scaling

# Generation per unit installed capacity (GWh per installed capacity unit)
PIC_ex1 = TE_cal1 / TP_ex1
PIC_ex2 = TE_cal2 / TP_ex2
PIC_sim = TE_sims / TP_sims

# Relative performance ratios (dimensionless)
RP1 = PIC_sim / PIC_ex1
RP2 = PIC_sim / PIC_ex2

# ---------------------------
# Firm energy (5th percentile) and gains
# Compute 5th-percentile daily power across time for each scenario
firm_opt  = np.quantile(Sim_DailyPowers, 0.05, axis=0)      # HyTUNE firm daily power
firm_fix1 = np.quantile(Cal_DailyPowers1, 0.05, axis=0)     # R1 firm daily power
firm_fix2 = np.quantile(Cal_DailyPowers2, 0.05, axis=0)     # R2 firm daily power

# Firm energy gain: HyTUNE minus fixed. Original code scaled by /1000.
firm_gain1 = (firm_opt - firm_fix1) / 1000.0
firm_gain2 = (firm_opt - firm_fix2) / 1000.0

# ---------------------------
# Peak power comparisons
R1_PeakEnergy = np.loadtxt(INPUT_DIR /"ALL1R_Cal_DailyPowersPeak.csv", delimiter=",")
R2_PeakEnergy = np.loadtxt(INPUT_DIR /"ALL2R_Cal_DailyPowersPeak.csv", delimiter=",")

# Per-day peak power improvement (HyTUNE - benchmark)
Peak_power1 = Sim_PeakPowers - R1_PeakEnergy
Peak_power2 = Sim_PeakPowers - R2_PeakEnergy

# Mean annualized peak-power gain per scenario (GWh)
Peak_power1_mean = Peak_power1.mean(axis=0) * 365 / 1e6
Peak_power2_mean = Peak_power2.mean(axis=0) * 365 / 1e6

# ---------------------------
# Economic comparison: NPV gain (HyTUNE minus benchmark)
dif_NPV1R = Sim_of - NPV1
dif_NPV2R = Sim_of - NPV2
```
